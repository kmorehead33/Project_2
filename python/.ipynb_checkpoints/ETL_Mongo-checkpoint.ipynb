```python
import pandas as pd
import numpy as np
import pymongo
import pprint
```

```python
allstar_db = pd.read_csv("../data/AllstarFull.csv")
```

```python
allstar_db = allstar_db[['yearID', 'teamID', 'playerID']]
allstar_db = allstar_db.rename(columns={'yearID':'year', 'teamID':'team'})
```

```python
teams_db = pd.read_csv("../data/Teams.csv")
```

```python
teams_db = teams_db[['yearID', 'teamID', 'W', 'L', 'WSWin', 'attendance']]
teams_db = teams_db.rename(columns={'yearID':'year', 'teamID':'team', 'W':'Wins', 'L':'Losses', 'WSWin': 'Championship'})
```

```python
teams_db.tail()
```

```python
salaries_db = pd.read_csv("../data/Salaries.csv")
```

```python
salaries_db = salaries_db[['yearID', 'teamID', 'salary']]
salaries_db = salaries_db.rename(columns={'yearID': 'year', 'teamID':'team'})
```

```python
before_1985 = teams_db['year']>=1985
```

```python
teams_db = teams_db[before_1985]
```

```python
allstar_1985 = allstar_db['year']>=1985
allstar_db = allstar_db[allstar_1985]
```

```python
team_salary_db = pd.merge(teams_db, salaries_db, on=['year','team'])
team_salary_db
```
