```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
baseball_db = client['baseball_db']
```

```python
stats = baseball_db['stats']
```

```python
team_salary_list = team_salary_db.values.tolist()
team_salary_list
```

```python
baseball_dict = []
for list in team_salary_list:
    dictionary={}
    dictionary['year']=list[0]
    dictionary['team']=list[1]
    dictionary['wins']=list[2]
    dictionary['losses']=list[3]
    dictionary['championship']=list[4]
    dictionary['attendance']=list[5]
    dictionary['salary']=list[6]
    baseball_dict.append(dictionary)
    stats.insert_one(dictionary)
    
```
